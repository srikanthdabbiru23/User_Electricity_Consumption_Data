```python
## CONVERTING CATEGORICAL FEATURES
```

```python
kind = pd.get_dummies(df['TYPE'],drop_first=True)
start = pd.get_dummies(df['START TIME'],drop_first=True)
end = pd.get_dummies(df['END TIME'],drop_first=True)
units = pd.get_dummies(df['UNITS'],drop_first=True)
notes = pd.get_dummies(df['NOTES'],drop_first=True)
date = pd.get_dummies(df['DATE'],drop_first=True)
```

```python
df.drop(['TYPE','START TIME','END TIME','UNITS', 'DATE', 'NOTES'],axis=1)
df = pd.concat([kind,start,end,units,notes,date],axis=1)
```

```python
## Notice the Logistic Regression Model execution, along with the Classification report
```

```python
from sklearn.cross_validation import train_test_split
```

```python
X = df.drop(str('USAGE'), axis = 1)
```

```python
y = df[str('USAGE')]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```
